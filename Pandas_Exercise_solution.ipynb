```python
#Question 6: Find each company’s Higesht price car
# Read the CSV file, group data by 'company', find the maximum price for each company, and print the result
import pandas as pd

df = pd.read_csv("/content/Automobile_data.csv")

car_Manufacturers = df.groupby('company')

priceDf = car_Manufacturers[['company', 'price']].max()

print(priceDf)
```

```python
#Question 10: Merge two data frames using following condition
   # Create two data frames using following two Dicts, Merge two data frames, and append second data frame as a new column to first data frame.

Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

```python
import pandas as pd
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
carPriceDf = pd.DataFrame.from_dict(Car_Price)
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
carsHorsepowerDf = pd.DataFrame.from_dict(car_Horsepower)
carsDf = pd.merge(carPriceDf, carsHorsepowerDf, on="Company")
carsDf
```
