[1,2,3,4,5] → These are your x-values.

[1,4,9,16,25] → These are your y-values (squares of x-values).

'r^-' → This is a format string that styles the plot:

'r' → Red color

'^' → Triangle-up markers for each point

'-' → Solid line connecting the points

```python
plt.grid(True)
plt.title('squares of the numbers')
plt.xlabel('The Numbers', fontsize=18, color='red')
plt.ylabel('The squares',fontsize=18, color='red')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'r^--', linewidth=3.0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Import the 3D plotting toolkit
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D Axes object
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Sample data for a 3D surface plot with a new shape (saddle)
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)

# New function for Z to create a saddle shape
Z = X**2 - Y**2 # This creates a classic saddle shape

# Create a 3D surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none') # Changed cmap for variety

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Saddle Surface Plot with Heatmap Coloring')

# Add a color bar which maps values to colors
fig.colorbar(surf, shrink=0.5, aspect=5, label='Z-value (Intensity)')

# Show the plot
plt.show()

# saving the figure
plt.savefig('saddle_surface_plot.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a set of 3D subplots arranged in a 2x2 grid
fig = plt.figure(figsize=(15, 12))

# --- Plot 1: 3D Surface Plot (Similar to previous Z = sin(R)) ---
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
X_surf = np.arange(-5, 5, 0.25)
Y_surf = np.arange(-5, 5, 0.25)
X_surf, Y_surf = np.meshgrid(X_surf, Y_surf)
R_surf = np.sqrt(X_surf**2 + Y_surf**2)
Z_surf = np.sin(R_surf)
ax1.plot_surface(X_surf, Y_surf, Z_surf, cmap='viridis')
ax1.set_title('3D Surface Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# --- Plot 2: 3D Scatter Plot (Random data) ---
ax2 = fig.add_subplot(2, 2, 2, projection='3d')
num_points = 50
X_scatter = np.random.rand(num_points) * 10
Y_scatter = np.random.rand(num_points) * 10
Z_scatter = np.random.rand(num_points) * 10
ax2.scatter(X_scatter, Y_scatter, Z_scatter, c=Z_scatter, cmap='plasma', marker='o')
ax2.set_title('3D Scatter Plot')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

# --- Plot 3: 3D Bar Plot ---
ax3 = fig.add_subplot(2, 2, 3, projection='3d')
xpos = np.arange(4)
ypos = np.arange(4)
xpos, ypos = np.meshgrid(xpos, ypos)
xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like(xpos)
dx = np.ones_like(xpos) * 0.5
dy = np.ones_like(ypos) * 0.5
dz = np.random.rand(len(xpos)) * 5 # Heights of the bars
ax3.bar3d(xpos, ypos, zpos, dx, dy, dz, color='cyan', alpha=0.8)
ax3.set_title('3D Bar Plot')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')

# --- Plot 4: 3D Wireframe Plot ---
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
X_wire = np.linspace(-5, 5, 20)
Y_wire = np.linspace(-5, 5, 20)
X_wire, Y_wire = np.meshgrid(X_wire, Y_wire)
Z_wire = np.cos(np.sqrt(X_wire**2 + Y_wire**2)) * 3
ax4.plot_wireframe(X_wire, Y_wire, Z_wire, rstride=1, cstride=1, color='green')
ax4.set_title('3D Wireframe Plot')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.set_zlabel('Z')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')

# Create a grid of points
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Define the direction of the vectors (e.g., a rotating field)
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) * np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True, cmap='inferno')

ax.set_title("3D Quiver Plot (Vector Field)")
plt.show()
```

```python
# Unique 3D Spiral Surface Plot in Python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create figure
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Parameters
theta = np.linspace(0, 8 * np.pi, 300)
z = np.linspace(-2, 2, 300)

Theta, Z = np.meshgrid(theta, z)

# Spiral radius
R = Z**2 + 1

# Parametric equations
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Plot surface
surf = ax.plot_surface(
    X, Y, Z,
    cmap='twilight',
    edgecolor='none'
)

# Add color bar
fig.colorbar(surf, shrink=0.5, aspect=10)

# Labels and title
ax.set_title("Unique 3D Spiral Surface")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Stylish viewing angle
ax.view_init(elev=35, azim=60)

plt.show()
```
